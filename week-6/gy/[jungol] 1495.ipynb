# 대각선 지그재그

https://jungol.co.kr/problem/1495?cursor=eyJwcm9ibGVtc2V0Ijo2LCJmaWVsZCI6MiwiaWR4IjoxMn0%3D

- Solve idea
  - arr의 왼쪽 위는 1이다.
  - 첫번째로 내려간다 (한 번만)
    - 이동을 못하면 세번째를 실행
  - 두번째로 대각선 위로 간다 (끝까지)
    - 다 도달하면 세번째를 실행
  - 세번쨰로 오른쪽으로 이동한다 (한 번만)
    - 이동을 못하면 끝
  - 네번째로 대각선 밑으로 간다. (끝까지)
    - 다 도달하면 첫번째를 실행

```python
import sys
input = sys.stdin.readline
n = int(input())
arr = [[0]*(n) for _ in range(n)]
arr[0][0] = 1
def first_step(x,y,ptr,flag):
    if x+1 < n:
        x += 1
    else:
        
        return 3, x, y, ptr, flag
        
    arr[x][y] = ptr
    ptr += 1
    if flag: # 절반 넘으면
        
        return 4, x, y, ptr, flag
    else:
        return 2, x, y, ptr, flag
        
def second_step(x,y,ptr,flag):
    cnt = 0
    while True:
        if 0<= x-1 < n and 0<= y + 1 < n:
            x -= 1
            y += 1
            
            arr[x][y] = ptr
            ptr += 1
        else:
            break
        cnt += 1
    if cnt == n-1:
        
        flag = 1
    return 3, x, y, ptr, flag
            
def third_step(x,y,ptr,flag):
    if y+1 <n:
        y += 1

    else:
        return 1, x, y, ptr, flag
    arr[x][y] = ptr
    ptr += 1
    if flag: # 절반 넘으면
        return 2, x, y, ptr, flag
    else:
        return 4, x, y, ptr, flag
        
def fourth_step(x,y,ptr,flag):
    cnt = 0
    
    while True:
        if x + 1 < n and  0<= y-1 <n:
            x += 1
            y -= 1
            
            arr[x][y] = ptr
            ptr += 1
        else:
            break
        cnt += 1
    if cnt == n-1:
        
        flag = 1
    return 1, x, y, ptr, flag
        
ptr = 2
x,y = 0,0
mode = 1
mode2 = 0
while True:
    
    if x == n-1 and y == n-1:
        break
        
    if mode == 1:
        mode,x,y,ptr, mode2 = first_step(x,y,ptr, mode2)
        
    elif mode == 2:
        mode,x,y,ptr, mode2 = second_step(x,y,ptr, mode2)
        
    elif mode == 3:
        mode,x,y,ptr, mode2 = third_step(x,y,ptr, mode2)
        
    elif mode == 4:
        mode,x,y,ptr, mode2 = fourth_step(x,y,ptr, mode2)
        
    else: # 끝
        break
for i in arr:
    print(*i)
```
